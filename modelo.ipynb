```python
import pandas as pd
import selenium

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://www.fundamentus.com.br/resultado.php"

driver.get(url)
```

```python
local_tabela = "/html/body/div[1]/div[2]/table"
elemento = driver.find_element("xpath", local_tabela)
tabela_html = elemento.get_attribute("outerHTML")

dataf = pd.read_html(str(tabela_html), thousands = ".", decimal = ",")[0]
dataf
```

```python
dataf = dataf.set_index("Papel")
dataf = dataf[["Cotação", "EV/EBIT", "ROIC", "Liq.2meses"]]
dataf
```

```python
dataf.info()

dataf["ROIC"] = dataf["ROIC"].str.replace("%", "")
dataf["ROIC"] = dataf["ROIC"].str.replace(".", "")
dataf["ROIC"] = dataf["ROIC"].str.replace(",", ".")
dataf["ROIC"] = dataf["ROIC"].astype(float)
display(dataf)
```

```python
dataf = dataf[dataf["Liq.2meses"] > 1000000]
dataf = dataf[dataf["EV/EBIT"] > 0]
dataf = dataf[dataf["ROIC"] > 0 ]
```

```python
dataf["ranking_ev_ebit"] = dataf["EV/EBIT"].rank(ascending= True)
dataf["ranking_roic"] = dataf["ROIC"].rank(ascending= False)
dataf["ranking_total"] = dataf["ranking_ev_ebit"] + dataf["ranking_roic"]

dataf = dataf.sort_values("ranking_total").rank(ascending= True).astype(int)

dataf.head(10)

```
